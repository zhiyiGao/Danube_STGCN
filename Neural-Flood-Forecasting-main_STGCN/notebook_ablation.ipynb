```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
import pandas as pd
```

```python
results = pd.read_csv("/home/kirschstein/floodgnn/results/results_ablation.csv", index_col=0)

def extract_results(window_size, lead_time):
    return results.loc[[f"{window_size}_{lead_time}_{fold}" for fold in range(3)]]

def print_table():
    display(pd.DataFrame(
        [
            sum([["{:.2f}%".format(100 * extract_results(window_size, lead_time).mean(1).mean())
                    + " ± {:.2f}%".format(100 * extract_results(window_size, lead_time).mean(1).std())]
             for lead_time in [1, 2, 3, 6, 9, 12]], [])
           for window_size in [12, 24, 36, 48, 60, 72]
        ], 
        columns = ["1", "2", "3", "6", "9", "12"],
        index = ["12", "24", "36", "48", "60", "72"]
    ))
```

```python
print_table()
```
